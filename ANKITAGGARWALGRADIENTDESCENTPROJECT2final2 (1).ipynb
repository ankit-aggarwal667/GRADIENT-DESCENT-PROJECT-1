```python
import numpy as np
def gd(data,learning_rate,number_of_iteration,n):
    m=[0 for i in range(n)]
    c=0
    for i in range(number_of_iteration):
        m,c=step_gradient(data,learning_rate,m,c,n)
        cost(data,m,c)
    return m,c
def cost(data,m,c):
    total=0
    M=len(data)
    for i in range(M):
        x=data[i,0:-1]
        y=data[i,-1]
        val=0
        for j in range(len(x)):
            val+=m[j]*x[j]
        total+=(1/M)*((y-val-c)**2)
    return total
def step_gradient(data,learning_rate,m,c,n):
    m_slope=[0 for i in range(n)]
    c_slope=0
    M=len(data)
    for t in range(n):
        for i in range(M):
            x=data[i,0:-1]
            y=data[i,-1]
            val=0
            for j in range(len(x)):
                val+=m[j]*x[j]
            m_slope[t]+=(-2/M)*((y-val-c)*x[t])
            c_slope+=(-2/M)*(y-val-c)
    for i in range(n):
        m[i]=m[i]-learning_rate*m_slope[i]
    c=c-learning_rate*c_slope
    return m,c
def run():
    learning_rate=0.0099
    number_of_iteration=500
    m,c=gd(data1,learning_rate,number_of_iteration,13)
    return m,c
def predict(data,m,c,n):
    M=len(data)
    for i in range(M):
        x=data[i,:]
        val=0
        for j in range(n):
            val+=m[j]*x[j]
        ans=val+c
        y_pred.append(ans)

data1=np.loadtxt("C:/Users/Ankit Aggarwal/Documents/Gradient_descent_training_data.csv",delimiter=",")
data2=np.loadtxt("C:/Users/Ankit Aggarwal/Documents/Gradient_descent_test_data.csv",delimiter=",")
y_pred=[]
m,c=run()
predict(data2,m,c,13)
for i in y_pred:
    print(i)
np.savetxt("C:/Users/Ankit Aggarwal/Desktop/ankitaggarwalGradientdescentproject2final2.csv",y_pred,delimiter=",")
    
```
